# ðŸ”¬ Introduction to SciPy
This notebook demonstrates basic functionality of the SciPy library including optimization, integration, linear algebra, and statistics.

```python
import numpy as np
from scipy import stats, optimize, integrate, linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Curve Fitting

```python
x = np.linspace(0, 10, 100)
y = 3.5 * np.sin(x) + np.random.normal(size=100)
def model_func(x, a, b):
    return a * np.sin(b * x)

params, covariance = optimize.curve_fit(model_func, x, y)
print("Fitted parameters:", params)

plt.scatter(x, y, label='Data')
plt.plot(x, model_func(x, *params), color='red', label='Fitted curve')
plt.legend()
plt.title("Curve Fitting with SciPy")
plt.show()
```

## 2. Integration

```python
result, error = integrate.quad(lambda x: np.exp(-x**2), 0, np.inf)
print("Integral of exp(-x^2) from 0 to âˆž:", result)
```

## 3. Linear Algebra

```python
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x_sol = linalg.solve(A, b)
print("Solution to Ax = b:", x_sol)
```

## 4. Statistics

```python
group1 = np.random.normal(loc=5.0, scale=1.0, size=30)
group2 = np.random.normal(loc=5.5, scale=1.0, size=30)
t_stat, p_val = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat, "P-value:", p_val)
```

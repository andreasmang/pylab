# Visualizing 3D Objects in Python

This notebook visualizes a variety of 3D shapes using Python:
- A 3D Sphere
- A 3D Torus
- A MÃ¶bius Strip
- The Stanford Bunny (loaded from a local `.tar.gz` archive)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
import tarfile
import trimesh
```

## 1. 3D Sphere

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones_like(u), np.cos(v))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue', edgecolor='k', alpha=0.8)
ax.set_title("3D Sphere")
plt.show()
```

## 2. 3D Torus

```python
R = 1
r = 0.3
u, v = np.meshgrid(np.linspace(0, 2 * np.pi, 100),
                   np.linspace(0, 2 * np.pi, 100))
x = (R + r * np.cos(v)) * np.cos(u)
y = (R + r * np.cos(v)) * np.sin(u)
z = r * np.sin(v)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title("3D Torus")
plt.show()
```

## 3. MÃ¶bius Strip

```python
theta = np.linspace(0, 2 * np.pi, 100)
w = np.linspace(-0.3, 0.3, 20)
theta, w = np.meshgrid(theta, w)

R = 1
x = (R + w * np.cos(theta / 2)) * np.cos(theta)
y = (R + w * np.cos(theta / 2)) * np.sin(theta)
z = w * np.sin(theta / 2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='cornflowerblue', edgecolor='k', alpha=0.9)
ax.set_title("MÃ¶bius Strip")
plt.show()
```

## 4. Stanford Bunny
The Stanford Bunny is extracted from a local `.tar.gz` archive and rendered using `trimesh`.

```python
# Define the path to the .tar.gz archive
archive_path = "../data/bunny.tar.gz"
extracted_file = "bunny/bunny.ply"

# Extract if not already extracted
if not os.path.exists(extracted_file):
    with tarfile.open(archive_path, "r:gz") as tar:
        tar.extractall()

# Load and visualize the bunny
bunny = trimesh.load(extracted_file)
bunny.show()
```
